## Perason Corelation

Pearson correlation measures the linear relationship between two continuous
variables. The value of Pearson correlation ranges from -1 to +1. A value closer to +1
indicates a strong positive linear relationship, while a value closer to -1 indicates a strong
negative linear relationship.

```python
pearson_corr = df['Salary'].corr(df['Performance_Score'],
method='pearson')
print(f"Pearson Correlation between Salary and Performance Score:  {pearson_corr}")
```

```python
pearson_corr = df['Salary'].corr(df['Monthly_Hours_Worked'], method='pearson')
print(f"Pearson Correlation between Salary and Hours:  {pearson_corr}")
```

## Spearman Correlation

Spearman correlation is a test that measures the monotonic
relationship between two ranked variables. Unlike Pearson, it does not assume a linear
relationship. Spearman is used when the data doesn’t meet the assumptions required by
Pearson correlation.

```python
spearman_corr = df['Monthly_Hours_Worked'].corr(df['Salary'], method='spearman')
print(f"Spearman Correlation between Years with Company and Salary:  {spearman_corr}")
```

## Monotonic Vs Linear Relation

A linear relationship occurs when the change in one variable is proportional to the change in another variable, following a straight-line pattern.
Characteristics:
Represented by a straight line in a scatter plot.

Positive Linear Relationship: As one variable increases, the other increases at a constant rate.
Negative Linear Relationship: As one variable increases, the other decreases at a constant rate.
Example:
The distance traveled by a car is linearly related to time if the car is moving at a constant speed.


A monotonic relationship occurs when one variable consistently increases or decreases as the other variable changes, though not necessarily at a constant rate.
Characteristics:
The relationship can be increasing (both variables move in the same direction) or decreasing (one increases while the other decreases).
The change doesn't need to be uniform or follow a straight line.
Example:
Increasing Monotonic: As age increases, income may also increase, though not at a constant rate.




```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = df[['Salary', 'Performance_Score',
                         'Monthly_Hours_Worked']].corr(method='spearman')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.yticks(rotation=45)
plt.xticks(rotation=45)
plt.show()
```

## Step 1: Define Hypotheses and Generate Data

• Suppose we are testing whether the average monthly income in a city is $5,000.
• Null Hypothesis (H0): The average income is $5,000.
• Alternative Hypothesis (H1): The average income is not $5,000.


```python
import numpy as np
from scipy import stats
data = np.random.normal(loc=7030, scale=200, size=50)
print(data)
```

## Step2: Set Alpha Level

```python

alpha = 0.05
# Simulate some sample data (example monthly incomes)
```

## Step3:Conduct a one-sample t-test

```python

t_statistic, p_value = stats.ttest_1samp(data, 5000)
# Display results
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)


```

## Step4_ Compare P-value

## Step5_Interpret Results

```python

if (p_value < alpha):
  print("Reject the Null hypothesis")
  print("Salary Mean is not 5000")
else:
  print("Fail to reject the Null hypothesis")
  print("Salary Mean is 5000")
```
