```python
import pandas as pd

df = pd.DataFrame({
    'Name' : ['Ahsan', 'Latif', 'Hassan', 'Riaz', 'Iqra'],
    'Age': [23, 35, 50, 22, 150]
})
df
```

```python
df.sort_values(by='Age', ascending=True, inplace=True)
df
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 =df['Age'].quantile(0.75)

IQR = Q3 - Q1
IQR
```

```python
lower_boundry = Q1 - 1.5 * IQR
upper_boundry = Q3 + 1.5 * IQR

print(lower_boundry, upper_boundry)
```

```python
outliers = df[(df['Age']<lower_boundry) | (df['Age']>upper_boundry)]
outliers
```

```python
import numpy as np
df['Age']= np.where(df['Age']<lower_boundry, lower_boundry, df['Age'])
df['Age']= np.where(df['Age']>upper_boundry, upper_boundry, df['Age'])
df
```

```python
df['Age']= df['Age'].apply(lambda x: lower_boundry if x<lower_boundry else (upper_boundry if x> upper_boundry else x))
df
```
