
# Task 1: Loading the Dataset

In this task, we load the employee productivity dataset into a Pandas DataFrame to begin analyzing the data distributions.


```python

# Load the dataset
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/home/nak/Data-Science-Course-IBA/data/Employee_Productivity_Dataset.csv')
# Display the first few rows of the dataset
df.head()
```


# Task 2: Visualizing the Distribution of Salary

We use a histogram to visualize the distribution of the 'Salary' feature to understand its shape. The `sns.histplot()` function plots the histogram with a KDE (Kernel Density Estimate) for a smooth density curve.


```python

# Visualize the distribution of Salary using a histogram
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Salary'], kde=True, bins=30)
plt.title("Distribution of Salary")
plt.show()
```


# Task 3: Visualizing Monthly Hours Worked Using KDE

A KDE plot is used to visualize the distribution of the 'Monthly_Hours_Worked' feature. This plot shows the probability density of the data using a continuous curve.


```python

# Visualize the distribution of Monthly Hours Worked using KDE
sns.kdeplot(df['Monthly_Hours_Worked'], shade=True)
plt.title("KDE Plot for Monthly Hours Worked")
plt.show()
```


# Task 4: Identifying the Distribution Type of Performance Score

We plot both a histogram and a KDE plot for the 'Performance_Score' feature to visually identify its distribution type.


```python

# Visualize the distribution of Performance Score
sns.histplot(df['Performance_Score'], kde=True, bins=30)
plt.title("Distribution of Performance Score")
plt.show()
```


# Task 5: Calculating the Variance of Salary

Variance measures the spread or dispersion of data points around the mean. A higher variance indicates that the data points are more spread out. We calculate the variance of the 'Salary' feature using Pandas' `var()` function.


```python

# Calculate and display the variance of Salary
variance_salary = df['Salary'].var()
print(f"Variance of Salary: {variance_salary}")
```
