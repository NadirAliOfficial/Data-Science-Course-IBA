## Uniform Distribution
np.random.uniform(0, 100, 1000):
Generates 1000 random numbers from a uniform distribution.

The range is between 0 (inclusive) and 100 (exclusive).

A uniform distribution means every number in the range [0, 100) has an equal probability of being selected.


```python
uniform_data = np.random.uniform(0, 1, 1000)
plt.hist(uniform_data, bins=30, alpha=0.6, color='b')
plt.title("Uniform Distribution")
plt.show()
```

## Binomial Distribution

n = 10: Number of trials (repetitions of the experiment).

p = 0.5: Probability of success in each trial (like flipping a fair coin).

**binom.rvs(n, p, size=1000):**
Generates 1000 random samples from a Binomial distribution.

The distribution represents the number of successes in n=10 independent trials, where the probability of success is p=0.5.

```python
from scipy.stats import binom
n, p = 10, 0.5  # Number of trials and probability of success
binomial_data = binom.rvs(n=n, p=p, size=1000)
print(binomial_data)
```

```python
plt.hist(binomial_data, bins=30, density=True, alpha=0.6, color='r')
plt.title("Binomial Distribution")
plt.show()
```
