```python
# create a program. Give different options to user. String(all elments are string)
# 1. Append: if user enters 1 the program will ask the element that will be added to list
# 2. Delete: IT will delete the elemetn that is input by the user
# 3. print: this will print all the lements in the list
# 4. Sort: it will print the sorted list


ch = int(input("input your choice: 1. Append 2. Delete, 3. Print, 4. Sort"))
if ch == 1:
  x = input("Enter element to append: ")
  names.append(x)
  print(names)
elif ch == 2:
  x = input("Enter element to delete: ")
  if x in names:
    names.remove(x)
    print(names)
  else:
    print("Not found")
elif ch == 3:
  print(names)
elif ch == 4:
  sorted = names.sort()
  print()
else:
  print("INvalid Choice")
```
