```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob',
'Charlie']})
df1.head()
```

```python
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Score': [85, 90, 95]})
df2.head()
```

```python
# Merge the two DataFrames on 'ID'
df_merged = pd.merge(df1, df2, on='ID', how='inner')
print(df_merged)
```

```python
import numpy as np
# Create a DataFrame with missing values
data_with_nan = {
'Name': ['John', 'Sara', 'Michael', 'Sophia', 'David'],
'Age': [23, 35, np.nan, 22, 32],
'City': ['New York', 'Los Angeles', 'Chicago', np.nan, 'Phoenix']
}
df3 = pd.DataFrame(data_with_nan)
df3
```

```python
df4 = pd.DataFrame({'Name': ['Daniel', 'Emily'], 'Age': [27, 30]})
df4
```

```python
# Concatenate DataFrames vertically
df_concat_vert = pd.concat([df3, df4], ignore_index=True)
print("Concatenated Vertically:\n", df_concat_vert)
```
