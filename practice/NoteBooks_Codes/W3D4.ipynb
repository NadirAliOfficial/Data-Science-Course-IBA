```python
import matplotlib.pyplot as plt

month = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'August', 'September', 'Oct', 'Nov', 'Dec']
sales = [2100, 2200, 3000, 2500, 2700, 3200, 3500, 3700, 4000, 4200, 4300, 4400]

plt.plot(month, sales, color='red', linestyle= '--', marker ='D')
plt.xlabel('Months')
plt.ylabel('Sales of othe month')
plt.title('Month Vs sales graph')
plt.grid(True)
plt.show()
```

```python
Department = ['Adminitration', 'Finance', 'Marketing', 'Procurement', 'HR', 'Transport']
Salary = [2000, 1800, 2600, 1800, 2500, 2000]

plt.plot(Department, Salary, color= 'red', linestyle='--', marker='x')
plt.xlabel ('Departments')
plt.ylabel('Salary')
plt.title('Salaries of Departments')
plt.show()
```

```python
Department = ['Adminitration', 'Finance', 'Marketing', 'Procurement', 'HR', 'Transport']
Salary = [2000, 1800, 2600, 1800, 2500, 2000]

plt.bar(Department, Salary, color='green', edgecolor= 'black')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.title('Bar chart of salaries versus department')
plt.show()
```

```python
Department = ['Adminitration', 'Finance', 'Marketing', 'Procurement', 'HR', 'Transport']
Salary = [2000, 1800, 2600, 1800, 2500, 2000]

plt.scatter(Department, Salary)
plt.xlabel('Department')
plt.ylabel('Salary')
plt.title('Scatter plot of salaries versus department')
plt.show()
```

```python
Department = ['Adminitration', 'Finance', 'Marketing', 'Procurement', 'HR', 'Transport']
Salary = [2000, 1800, 2600, 1800, 2500, 2000]
Anual_sale = [ 1800, 2000, 2200, 1900, 2000, 2100]

plt.figure(figsize=(4,5))
plt.plot(Department, Salary , color = 'red', marker='o', label='Department Salary')
hold_plot = plt.gca()
hold_plot.set_ylim(1500,3000)
hold_plot.set_xlim('Adminitration', 'HR' )
plt.plot(Department, Anual_sale, color= 'green', marker='s', label= 'Annual Sale')
plt.legend()
plt.xlabel('Department')
plt.ylabel('Salary')
plt.title('Scatter plot of salaries versus department')
plt.show()
```

```python
Department = ['Adminitration', 'Finance', 'Marketing', 'Procurement', 'HR', 'Transport']
Salary = [2000, 1800, 2600, 1800, 2500, 2000]

plt.barh(Department, Salary, color= ['green', 'blue'] , edgecolor= 'black')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.title('Bar chart of salaries versus department')
plt.savefig('barchart.png')
plt.show()
```
