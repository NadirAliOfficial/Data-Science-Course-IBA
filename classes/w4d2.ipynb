Task 1: Introduction to Normal Distribution

A normal distribution, also known as the Gaussian distribution, is symmetric and follows
the bell curve.<br> It is one of the most common distributions in statistics and data analysis.
Generate a dataset that follows a normal distribution and visualize it.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(0)

# Generate normal data
normal_data = np.random.normal(loc=50, scale=10, size=1000)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot histogram with seaborn
sns.histplot(normal_data, bins=30, kde=True, ax=ax, alpha=0.6, color='g')

# Set title and labels
ax.set_title("Normal Distribution")
ax.set_xlabel("Value")
ax.set_ylabel("Density")

# Show the plot
plt.tight_layout()
plt.show()
```

Task 2: Introduction to Uniform Distribution

A uniform distribution has all values equally likely to occur within a range. Generate and
visualize a uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

uniform_data = np.random.uniform(0, 1, 1000)
plt.hist(uniform_data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

Task 3: Introduction to Binomial Distribution
A binomial distribution is used to model the number of successes in a fixed number of
trials, with each trial having two possible outcomes. Generate and visualize a binomial
distribution.

```python
from scipy.stats import binom
n, p = 10, 0.5 # Number of trials and probability of success
binomial_data = binom.rvs(n=n, p=p, size=1000)
plt.hist(binomial_data, bins=30, density=True, alpha=0.6, color='r')
plt.title("Binomial Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Step 1: Create a population of customer spending
population_size = 10000
population_spending = np.random.normal(loc=500, scale=100,
size=population_size)
# Step 2: Take 100 random samples of size 200 each
sample_size = 200
n_samples = 100
sample_means = []
for _ in range(n_samples):
    sample = np.random.choice(population_spending, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))
# Step 3: Plot the sampling distribution of the sample means
plt.hist(sample_means, bins=20, edgecolor='black', alpha=0.7)
plt.title('Sampling Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
