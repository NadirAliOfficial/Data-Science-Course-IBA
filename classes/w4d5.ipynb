```python
#  Supervised learning and Unsupervised learning

#  - Supervised learning: The algorithm is trained on labeled data.
#  - Unsupervised learning: The algorithm is trained on unlabeled data.
#  - Semi-supervised learning: The algorithm is trained on both labeled and unlabeled data.
```

```python
# Classifications and Regression
# - Classifications: The output is a categorical variable...
# - Regression: The output is a continuous variable...

```

```python
import pandas as pd

data = pd.read_csv('../data/Housing.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
X = df[['area']]
Y = df[['price']]
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_test, Y_test, color='blue', label='Actual Data')
plt.plot(X_test, y_pred, color='g', label='Predicted Data')
plt.legend()
plt.show()
```

```python
print("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X = df[['area', 'bedrooms', 'stories']]
Y = df[['price']]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, Y_train)

y_pred = model.predict(X_test)

plt.scatter(Y_test, y_pred, color='blue', label='Actual Data')
plt.plot(Y_test, Y_test, color='g', label='Prediction')

# plt.scatter(X_test, Y_test, color='blue', label='Actual Data')
# plt.plot(X_test, y_pred, color='g', label='Predicted Data')

plt.legend()
plt.show()

print("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)
```
