```python
import pandas as pd
data = pd.read_csv("../data/employees_data.csv")
df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.bar(df.Department, df.Salary)
plt.xlabel("Departments")
plt.ylabel("Salaries")
plt.title("Salaries of Departments")
plt.show()
```

```python
maximum = df.groupby("Department")["Salary"].max()
minimum = df.groupby("Department")["Salary"].min()
plt.bar(maximum.index, maximum)
plt.bar(minimum.index, minimum)
plt.xlabel("Departments")
plt.ylabel("Salaries")
plt.title("Salaries of Departments")
plt.legend(["Maximum Salary", "Minimum Salary"])
plt.show()
```

```python
department_count = df.Department.value_counts()
department_count
```

```python
plt.pie(department_count, labels=department_count.index, autopct="%1.0f%%", explode=[0,0.1,0,0.1,0,0.1])
plt.show()
```

```python
plt.hist(df.Salary, bins=10, color='green', edgecolor='black')
plt.xlabel("Salaries")
plt.ylabel("Frequency")
plt.title("Histogram of Salaries")
plt.show()
```

```python
df["Join_Date"] = pd.to_datetime(df["Join_Date"])
trend = df["Join_Date"].dt.year.value_counts().sort_index()

people = trend
plt.plot(people.index, people, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Employees')
plt.title('Number of Employees Joined Each Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='Department', y='Salary', data=df)
plt.show()
```
