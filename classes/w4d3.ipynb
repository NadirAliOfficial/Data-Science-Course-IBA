```python
#  Today We are going to learn a new thing:
import pandas as pd
data = pd.read_csv("../data/Employee_Productivity_Dataset.csv")
df = pd.DataFrame(data)
df.head()
```

Monotonic Vs Linear Relation
A linear relationship occurs when the change in one variable is proportional to the change in another variable, following a straight-line pattern. Characteristics: Represented by a straight line in a scatter plot.

Positive Linear Relationship: As one variable increases, the other increases at a constant rate. Negative Linear Relationship: As one variable increases, the other decreases at a constant rate. Example: The distance traveled by a car is linearly related to time if the car is moving at a constant speed.

A monotonic relationship occurs when one variable consistently increases or decreases as the other variable changes, though not necessarily at a constant rate. Characteristics: The relationship can be increasing (both variables move in the same direction) or decreasing (one increases while the other decreases). The change doesn't need to be uniform or follow a straight line. Example: Increasing Monotonic: As age increases, income may also increase, though not at a constant rate.

```python
pearson_corr = df['Salary'].corr(df['Performance_Score'],
method='pearson')
print(f"Pearson Correlation between Salary and Performance Score: {pearson_corr}")
```

```python
spearman_corr = df['Monthly_Hours_Worked'].corr(df['Salary'], method='spearman')
print(f"Spearman Correlation between Years with Company and Salary: {spearman_corr:.4f}")
```

```python
import numpy as np
from scipy.stats import norm
# Known population standard deviation (sigma)
sigma = 8000
# Sample data: Assume df['Salary'] is our sample column
sample_mean = np.mean(df['Salary'])
population_mean = 50000
n = len(df['Salary'])
# Calculate the z-statistic
z_stat = (sample_mean - population_mean) / (sigma / np.sqrt(n))
print(f"Z-statistic: {z_stat}")
# Calculate the p-value from the z-statistic (for a two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print(f"P-value: {p_value}")
```

```python
import numpy as np
from scipy import stats

# Simulate some sample data (example monthly incomes)
data = np.random.normal(loc=5050, scale=200, size=50)

# Conduct a one-sample t-test
t_statistic, p_value = stats.ttest_1samp(data, 5000)

# Display results
print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
