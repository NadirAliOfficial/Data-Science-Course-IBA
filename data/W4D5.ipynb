```python
import pandas as pd
```

```python
#loading data
df=pd.read_csv('/content/drive/MyDrive/Datasets/Housing.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
X= df[['area']]
Y =df[['price']]
```

```python
print(X,Y)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tarin, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
x_tarin.shape, x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_tarin, y_train)
```

```python
y_pred = model.predict(x_test)
x_test.head(3), y_pred[:3], y_test[:3]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test, y_test, color='red', label= 'Actual Data')
plt.plot(x_test, y_pred, color='green', label= 'Predicted Data')
plt.legend()
plt.show()
```

```python
model.coef_
```

```python
model.intercept_
```
